```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
df1 = pd.read_csv('Resources/athletic_sales_2020.csv')
df2 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
df1 = pd.read_csv('Resources/athletic_sales_2020.csv')
df1.head(-5)
```

```python
# Display the 2021 sales DataFrame
df2 = pd.read_csv('Resources/athletic_sales_2021.csv')
df2.head(-5)
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
type(df1)
```

```python
# Check the 2021 sales data types.
type(df2)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
df1_df2 = pd.concat([df1, df2], axis="rows", join="inner").reset_index(drop=True)
df1_df2
```

```python
# Check if any values are null.
df1_df2.isnull().any()
```

```python
# Check the data type of each column
df1_df2.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
df1_df2['invoice_date'] = pd.to_datetime(df1_df2['invoice_date'], errors = "coerce")
df1_df2.head()
```

```python
# Confirm that the "invoice_date" data type has been changed.
df1_df2['invoice_date']
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

grouped_data = df1_df2.groupby(["region", "state", "city"])['units_sold'].agg(Total_Products_Sold=("sum"))




# Show the top 5 results.
grouped_data.sort_values(by=(['Total_Products_Sold']), ascending = False).head(5)
```

```python
#grouped_data
```

#### Using `pivot_table`

```python
# Create a pivot table to show the number of products sold for region, state, and city
pt_products_sold = grouped_data.pivot_table(index=['region', 'state', 'city'], values='units_sold', aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
pt_products_sold.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 result
top_5_results = pt_products_sold.nlargest(5, 'Total_Products_Sold')
top_5_results
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_sales = df1_df2.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()

# Show the top 5 results.
top_5_results = grouped_sales.nlargest(5, 'total_sales')
top_5_results
```

#### Using `pivot_table`

```python
#Create a pivot table to show total sales for each region, state, and city
pivot_table_sales = df1_df2.pivot_table(index=["region", "state", "city"], values="total_sales", aggfunc="sum")

# Rename the "total_sales" column to "Total Sales"
pivot_table_sales.rename(columns={"total_sales": "Total Sales"}, inplace=True)

# Sort the results in descending order to show the top 5 regions, states, and cities with the greatest number of products sold
top_5_results = pivot_table_sales.sort_values(by="Total Sales", ascending=False).head(5)

# Display the top 5 results
top_5_results
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_sales = df1_df2.groupby(['retailer', 'region', 'state', 'city'])['total_sales'].sum().reset_index()

grouped_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
top_5_sales = grouped_sales.nlargest(5, 'Total Sales')
top_5_sales
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
pivot_table_sales = pd.pivot_table(df1_df2, 
                                   values='total_sales', 
                                   index=['retailer', 'region', 'state', 'city'], 
                                   aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
top_5_sales = pivot_table_sales.nlargest(5, 'Total Sales')

top_5_sales
```
